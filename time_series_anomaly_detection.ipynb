```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
logs = pd.read_csv('curriculum-access.txt', 
                    engine='python',
                    header=None,
                    index_col=False,
                    names = ['date', 'time', 'page', 'user', 'cohort', 'IP'],
                    sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
                    na_values='"-"')
logs
```

```python
plt.figure(figsize=(16,9))
logs[logs.cohort == 55].groupby('date').page.nunique().plot()
```
